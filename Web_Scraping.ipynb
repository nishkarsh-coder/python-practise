```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://www.flipkart.com/search?q=realme+mobile"
```

```python
html = requests.get(url)
```

```python
html
```

```python
print(html.content)
print(type(html.content))
```

```python
print(html.text)
print(type(html.text))
```

```python
soup = BeautifulSoup(html.text, "html.parser")
```

```python
soup
```

```python
soup.prettify
```

```python
print(soup.title)
print(soup.title.text)
print(soup.title.text.split("|"))
```

```python
print(soup.div)  #yeh phela div dega...
print(soup.div.text)
```

```python
print(soup.find("div",class_="_4rR01T"))
print(soup.find("div",class_="_4rR01T").text)
```

```python
print(soup.find_all("div",class_="_4rR01T"))
mobiles=soup.find_all("div",class_="_4rR01T")
```

```python
mobile=[]
mobile_colour=[]
mobile_storage=[]
for i in mobiles:
  print(i.text)
  mobile.append(i.text.split(" (")[0])
  x= i.text.split("(")[1]
  mobile_colour.append(x.split(",")[0])
  mobile_storage.append(x.split(",")[1].split(")")[0])
```

```python
# list2=soup.find_all("li",class_="rgWa7D")
# list2[0].text.split("|")[0]
# list2
list2=[]
list_ul= soup.find_all("ul",class_="_1xgFaf")
for i in list_ul:
  list2.append(i.find("li",class_="rgWa7D").text)
print(len(list2))
list2
```

```python
print(mobile_colour)
print(mobile_colour)
print(mobile_storage)
```

```python
ram=[]
for i in list2:
  ram.append(i.split("|")[0])
print(ram)
```

```python
mobile_price=[]
mobile_prices= soup.find_all("div",class_="_30jeq3 _1_WHN1")
for i in mobile_prices:
  mobile_price.append(i.text)
```

```python
import pandas as pd
dict1={"mobile" : mobile,"mobile_colour": mobile_colour,"mobile_storage": mobile_storage,"mobile_ram":ram,"mobile_price":mobile_price}
df= pd.DataFrame(dict1)
df
```

```python
links = soup.find_all("a",class_="_1fQZEK")
links
```

```python
link=[]
for i in links:
  link.append("https://www.flipkart.com"+i["href"].split("?")[0])
link
```
